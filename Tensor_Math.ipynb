```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# Reshape and View
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
my_torch2 = torch.arange(10)
my_torch2
```

```python
# Reshape if we don't know the number of items using -1
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
my_torch3 = torch.arange(15)
my_torch3
```

Differences between reshape and view

https://stackoverflow.com/questions/79025552/pytorch-difference-between-reshape-and-view-method
